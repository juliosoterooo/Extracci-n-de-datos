**a) Eliminar la fila “0” de “Unnamed” y Eliminar la columna “0” y la columna “Referencia”**

```python
# eliminar primera fila o filas innecesarias
df = pd.read_excel('Inmuebles.xlsx', skiprows=1)
df
```

```python
# eliminar columnas innecesarias
# axis = 1 para columnas
# axis = 0 para filas
df1 = df.drop(['Referencia'], axis=1)
df1
```

```python
df1.info()
```

**b) Todos las registros de “Fecha Alta : igual a 2007-04-29 00:00:00 y a 2004-01-02 00:00:00**

```python
df1['Fecha_Alta'] = pd.to_datetime(df1['Fecha_Alta'])

filtro2 = df1[(df1['Fecha_Alta'] == '2007-04-29 00:00:00') | (df1['Fecha_Alta'] == '2004-01-02 00:00:00')]
filtro2
#paso fecha alta a datetime para que lo lea como fecha
#luego le digo que de la columna sea igual a la primera fecha o a la otra
```

**c) Todos los registros de “ Tipo: parking, local, industrial y piso”**

```python
filtro3 = df1[(df1['Tipo'] == 'Parking') | (df1['Tipo'] == 'Local') | (df1['Tipo'] == 'Industrial') | (df1['Tipo'] == 'Piso')]
filtro3
#de la columna tipo las que sean parking o local o industrial o piso
```

**d) Todos los registros con “Provincia” igual a “Girona” y con “Superficie” mayor a “100”**

```python
filtro4= df1[(df1['Provincia'] == 'Girona') & (df1['Superficie'] > 100)]
filtro4
#que las de columna Provincia sea Girona Y con la columna Superficie mayor a 100
```

**e) Todos los registros con “Precio Venta: mayor a 1000,000” o “vendedor” igual a “Carmen o Joaquin”**

```python
filtro5= df1[(df1['Precio Venta'] > 1000000) | (df1['Vendedor'] == 'Carmen') | (df1['Vendedor'] == 'Joaquin')]
filtro5
#que en precio venta sea mayor a 1,000,000 o el vendedor sea carmen o joaquin
```

**f) Todos los registros que no sean de “Provincia : Barcelona“**

```python
filtro6 = df1[~(df1['Provincia'] == 'Barcelona')]
filtro6
#antes de df pongo ~ para que NO sean de Barcelona en la columna Provincia
```

**g)Todos los registros que no sean de “Operación: alquiler “ ni “Tipo: oficina“**

```python
filtro7 = df1[~((df1['Operación'] == 'Alquiler') | (df1['Tipo'] == 'Oficina'))]
filtro7
#~ antes del df para que no sean alquiler ni oficina de las columnas Operacion y Tipo
```
