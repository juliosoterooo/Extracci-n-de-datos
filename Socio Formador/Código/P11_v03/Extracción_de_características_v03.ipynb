```python
#Se genera el SubFrame con los datos donde cada emoción sea la establecida
meditacion = v03[v03['04_M']== 1]
meditacion.shape
```

```python
x='F5EEX'
```

```python
# Calculamos el numero total de la poblacion "n"
meditacion[x].info()
n_1 = len(meditacion)
#Obtenemos el limite superior y el limite inferior de la colmna objetivo
Max_1_1=meditacion[x].max()
Min_2_1=meditacion[x].min()
Limites= [Min_2_1, Max_1_1]
#Calculamos el Rango R
R_2_2=Max_1_1-Min_2_1
#Calculamos el numero de intervalos de clase "ni", aplicando la regla de Sturges
ni_1= 1+3.32*np.log10(n_1)
#Calculamos el Ancho del Intervalo "i"
i=R_2_2/ni_1
```

```python
Limites
```

```python
#Categorización de variables
intervalos=np.linspace(0.09565861,Max_1_1,round(ni_1))
(intervalos)
```

```python
#Creamos las categorias
categorias = ['0.09565861 - 0.14427617', '0.14427617 - 0.19289373', '0.19289373 - 0.24151130',
               '0.24151130 - 0.29012886', '0.29012886 - 0.33874642', '0.33874642 - 0.38736398',
               '0.38736398 - 0.43598154', '0.43598154 - 0.48459911', '0.48459911 - 0.53321667',
               '0.53321667 - 0.58183423', '0.58183423 - 0.63045179', '0.63045179 - 0.67906936',
               '0.67906936 - 0.72768692', '0.72768692 - 0.77630448']
```

```python
meditacion['F5EEX']=pd.cut(x=meditacion['F5EEX'],bins=intervalos,labels=categorias)
meditacion['F5EEX'].head()
```

Análisis univariado
