```python
#carga desde un archivo .xlsx sin indice
df = pd.read_excel('Datos de FacturacioÃÅn.xlsx')
#df
```

```python
df.info()
```

```python
df.head()
```

```python
#conversi√≥n de tipo de dato de columna de tipo object a float
df['CVE_CLPV'] = df['CVE_CLPV'].astype(int)
```

```python
#reemplazo de datos tipo string a datos tipo num√©rico
df.CVE_CLPV = df.CVE_CLPV.replace({'M':'0'}, regex=False)
```

**1-CVE_CLPV      (valores: 1000 a 2000)**

```python
filtro1 = df[(df['CVE_CLPV'] >= 1000) & (df['CVE_CLPV'] <= 2000)]
filtro1
#le indico que de la columna cve_clpv todo lo que sea mayor o igual a 1000
# Y todo lo que sea menor o igual a 2000
```

```python
pd.options.display.max_rows = 100
```

**2-CVE_VEND     (Todas las claves excepto ‚Äú5‚Äù y ‚Äú4‚Äù)**

```python
filtro2 = df[~(df["CVE_VEND"] == 5) & ~(df["CVE_VEND"] == 4)]
filtro2
#de la columna cve_vend las que no sean 5 y 4
```

**3-FECHA_ENT   (Fechas de ‚Äú28/02/2022‚Äù)**

```python
#conversi√≥n de tipo de dato de columna de tipo object a float
df['FECHA_ENT'] = df['FECHA_ENT'].astype(str)
```

```python
#analizar categorias de una columna
cats_fecha_ent = df.groupby(['CVE_CLPV'])['CVE_CLPV'].count()
cats_fecha_ent
```

```python
filtro3 = df[(df["FECHA_ENT"]== '2022-02-28 00:00:00')]
filtro3
```

**4-CAN_TOT       (Cantidades menores a 5951.7)   o  STATUS (‚ÄúE‚Äù)**

```python
filtro4 = df[(df["CAN_TOT"] < 5951.7) | (df["STATUS"] == "E")]
filtro4
#de la columna can_tot que sean menores a 5951 O de la columma status las que sean e
```

**5-Solo las columnas: CVE_DOC, FECHA_ENT, FECHA_VEN y CAN_TOT**

```python
filtro5 = df[['CVE_DOC', 'FECHA_ENT', 'FECHA_VEN', 'CAN_TOT']]
filtro5
#selecciono solo las columnas CVE_DOC, FECHA_ENT, FECHA_VEN y CAN_TOT
```

**6-Solo las filas de 7001-7099**

```python
filtro6 = df.iloc[7001:7099,:]
filtro6
#filas consecutivas de la 7001-7099
```

**7-Index= CVE_VEND (valores: 1, 2)  (columna: FECHAELAB)**

```python
df1 = pd.read_excel('Datos de FacturacioÃÅn.xlsx', index_col=3)
df1
#pongo de indice de columna a la 3, que es CVE_VEND
```

```python
filtro7 = df1.loc[[1,2], ["FECHAELAB"]]
filtro7
# 1 y 2 de la columna FECHAELAB
```
