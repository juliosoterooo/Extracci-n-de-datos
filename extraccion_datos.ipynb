**FILTROS BÁSICOS PARA VARIABLES CUALITATIVAS**

```python
#filtro por objeto
filtro1 = df[df["NOMBRE"] == "DONGA HWASUNG MEXICO"]
filtro1
```

```python
#filtro por filas
filtro2 = df.iloc[1:50,:]     #filas consecutivas
#filtro2 = df.iloc[ [8,24,100], :] #filas no consecutivas
filtro2
```

```python
#filtro por columnas
#filtro3 = df.iloc[ : , 1:3]     #columnas consecutivas
filtro3 = df.iloc[ : , [1,4,5]]   #columnas no consecutivas
filtro3
```

```python
df1 = pd.read_excel('Antigüedad de saldos 17.03.2023.xlsx', index_col=1)
df1
```

```python
#paso2: Aplicamos filtro de filas vs columnas
filtro4 = df1.loc[["HELVEX","DONGA HWASUNG MEXICO"], ["MONTO ADEUDADO"]]
filtro4
```

```python
#filtro de cabecera
filtro5 = df.tail()
filtro5
```

**FILTROS BASICOS PARA VARIABLES CUANTITATIVAS**\

```python
#filtro por comparación
filtro6 = df[df["MONTO ADEUDADO"] > 70000]
filtro6
```

**FILTROS BASICOS CON CONECTORES LÓGICOS**

```python
#filtro y
filtro7 = df[(df["MONTO ADEUDADO"] < 6000) & (df["FECHA_FACTURA"] == "2023-03-14")]
filtro7
```

```python
#filtro o
filtro8 = df[(df["MONTO ADEUDADO"] < 6000) | (df["FECHA_FACTURA"] == "2023-03-14")]
filtro8
```

```python
#filtro not
filtro9 = df[~(df["FECHA_FACTURA"] == "2023-03-14")]
filtro9
```
