**filtros para variables cualitativas**

```python
#filtro por objeto
filtro1 = df[df["NOMBRE_CLIENTE"] == "PÚBLICO EN GENERAL"]
filtro1
# aquí estoy buscando en la columna "NOMBRE_CLIENTE los que sean igual a "PÚBLICO EN GENERAL"
```

```python
#filtro por filas
#filtro2 = df.iloc[1:10,:]     #filas consecutivas
filtro2 = df.iloc[ [5,20,100], :] #filas no consecutivas
filtro2
# aquí llamo a las filas 5,20 y 100
```

```python
#filtro por columnas
#filtro3 = df.iloc[ : , 1:10]     #columnas consecutivas
filtro3 = df.iloc[ : , [1,5,8]]   #columnas no consecutivas
filtro3
#llamo a las columnas 1,5,8
```

```python
df1 = pd.read_excel('Detalle precios y productos fabricados 2022.xlsx', index_col=2)
df1
#pongo de indice de columna a la 2, que es nombre vendedor
```

```python
#aplicamos filtro de filas vs columnas
filtro4 = df1.loc[["LETICIA RAMIREZ HERNANDEZ","AARON NUÑO CORDOVA"], ["NOMBRE_CLIENTE"]]
filtro4
#llamo a las filas que son leticia y aaron y a la columna nombre cliente
```

```python
#filtro de cabecera
filtro5 = df.tail()
filtro5
#imprime los ultimos 5 registros
```

**FILTROS BASICOS PARA VARIABLES CUANTITATIVAS**

```python
#filtro por comparación
filtro6 = df[df["COSTO_UNITARIO"] > 10000]
filtro6
#de la columna costo unitario los que sean mayores a 10,000
```

**FILTROS BASICOS CON CONECTORES LÓGICOS**

```python
#filtro y
filtro7 = df[(df["PRECIO_UNITARIO"] < 1000) & (df["NOMBRE_CLIENTE"] == "PÚBLICO EN GENERAL")]
filtro7
# llamo a las columnas que su precio unitario sea menor que 1000 Y que en nombre cliente sea público en general
```

```python
#filtro o
filtro8 = df[(df["COSTO_UNITARIO"] < 500) | (df["PRECIO_UNITARIO"] < 500)]
filtro8
#llamo a las columnas que costo unitario son menores que 500 O precio unitario menor a 500
```

```python
#filtro not
filtro9 = df[~(df["CANT"] == 1) & ~(df["NOMBRE_VENDEDOR"] == "LETICIA RAMIREZ HERNANDEZ")]
filtro9
# de la columna CANT las que en cantidad NO sean 1 NI leticia ramirez hernandez
```
