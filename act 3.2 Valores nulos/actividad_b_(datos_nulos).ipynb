```python
data = pd.read_csv('Base_Socio_Formador_Con_Nulos.csv', encoding="latin")
data.head(5)
```

```python
data.info()
```

```python
#identificar valores nulos por columna
valores_nulos = data.isnull().sum()
valores_nulos
```

```python
#identificar valores nulos por dataframe
valores_nulos = data.isnull().sum().sum()
valores_nulos
```

```python
#realizamos una copia del dataframe
data2 = data.copy()
```

**fecha, año_hecho, año_renovado**

```python
#sustituir valores nulos por un string en concreto
data2["fecha"] = data2["fecha"].fillna("2030-06-19 00:00:00")
#sustituyo con fillna con una fecha porque es lo más sencillo
```

```python
valores_nulos = data2.isnull().sum()
valores_nulos
```

```python
data2["año_hecho"] = data2["año_hecho"].fillna(method="ffill")
data2["año_renovado"] = data2["año_renovado"].fillna(method="ffill")
#para evitar confusiones por fechas iguales a otras
```

```python
valores_nulos = data2.isnull().sum()
valores_nulos
```

**precio**

```python
data2.head(10)
```

```python
#data2.precio = data2.precio.replace({'$':''}, regex=False)
#Eliminar un signo de una columna   #guardar
data2['precio']=data2['precio'].str.replace('$', '')
```

```python
#Eliminar un signo de una columna
data2['precio']=data2['precio'].str.replace(',', '')
```

```python
data2['precio']= data2['precio'].astype(float)
```

```python
data2["precio"] = data2["precio"].fillna(round(data2["precio"].mean(),1))
#uso el promedio porque asi puedo encontrar una mejor distribución de los datos
```

```python
#corroboramos valores nulos
valores_nulos = data2.isnull().sum()
valores_nulos
```

**metros_cuadrados**

```python
data2["metros_cuadrados"] = data2["metros_cuadrados"].fillna(round(data2["metros_cuadrados"].mean(),1))
#uso el promedio porque asi puedo encontrar una mejor distribución de los datos
```

```python
#corroboramos valores nulos
valores_nulos = data2.isnull().sum()
valores_nulos
```

**pisos, habitaciones, baños**

```python
import statistics as st
```

```python
mode = st.mode(x for x in data['calle'] if x is not None)
mode
```

```python
mode = st.mode(x for x in data['codigo_zip'] if x is not None)
mode
```

```python
mode = st.mode(x for x in data['estado'] if x is not None)
mode
```

```python
mode = st.mode(x for x in data['ciudad'] if x is not None)
mode
```

```python
data2["calle"].fillna(mode, inplace=True)
data2["codigo_zip"].fillna(mode, inplace=True)
data2["estado"].fillna(mode, inplace=True)
data2["ciudad"].fillna(mode, inplace=True)
#uso la moda porque asi puedo encontrar el valor más repetido y una mejor
#representación de los datos
```

```python
valores_nulos = data2.isnull().sum()
valores_nulos
```

**pisos, habitaciones, baños**

```python
data2["pisos"] = data2["pisos"].fillna(round(data["pisos"].median(),1))
data2["habitaciones"] = data2["habitaciones"].fillna(round(data["habitaciones"].median(),1))
data2["baños"] = data2["baños"].fillna(round(data["baños"].median(),1))
#use mediana porque es un metodo ideal para datos atípicos porque es el punto medio
```

```python
valores_nulos = data2.isnull().sum()
valores_nulos
```
